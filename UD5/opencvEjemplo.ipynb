```python
import matplotlib.pyplot as plt
import cv2
```

```python
# Cargar la imagen en escala de grises
imagen = cv2.imread('corgi.jpeg', cv2.IMREAD_GRAYSCALE)
# Mostrar imagen con OpenCV (requiere ventana emergente)
cv2.imshow('Imagen', imagen)
cv2.waitKey(0)
cv2.destroyAllWindows()
# Mostrar con Matplotlib (evita problemas con colores)
plt.imshow(imagen, cmap='gray')
plt.axis('off')
plt.show()
```

```python
dim = (128, 128) #Tamao tpico para modelos como VGG16 o ResNet
imagen_redimensionada = cv2.resize(imagen, dim, interpolation=cv2.INTER_AREA)
```

```python
# Mostrar con Matplotlib (evita problemas con colores)
plt.imshow(imagen_redimensionada, cmap='gray')
plt.axis('off')
plt.show()
```

```python
imagen_redimensionada.shape
```

```python
print(imagen_redimensionada.max())
print(imagen_redimensionada.min())
```

```python
imagen_normalizada = imagen_redimensionada / 255.0
imagen_normalizada
```

```python
imagen_normalizada.max()
```

```python
imagen_normalizada.min()
```

```python
# Mostrar con Matplotlib (evita problemas con colores)
plt.imshow(imagen_normalizada, cmap='gray')
plt.axis('off')
plt.show()
```

```python
imagen_ecualizada = cv2.equalizeHist(imagen_normalizada)
```

```python
imagen_ecualizada2 = cv2.equalizeHist(imagen)
```

```python
# Mostrar con Matplotlib (evita problemas con colores)
plt.imshow(imagen_ecualizada2, cmap='gray')
plt.axis('off')
plt.show()
```
