```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
alumnado = ["Alumno1","Alumno2","Alumno3","Alumno4",]
mate = [1,2,7,8]
lengua = [3,5,8,9]
```

```python
plt.figure(figsize=(4, 2))
plt.plot(alumnado, mate)
plt.plot(alumnado, lengua)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,2))
axes[0].plot(alumnado, mate, color='red')
axes[1].plot(alumnado, lengua, color='blue')
plt.show()
```

```python
x = np.linspace(0, 180, 30)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(8, 4))
axes[0].plot(x, y1)
axes[0].set_title("Seno")
axes[1].plot(x, y2)
axes[1].set_title("Coseno")
axes[2].plot(x, y3)
axes[2].set_title("Tangente")
fig.suptitle("Funciones trigonomtricas")
plt.show()
```
